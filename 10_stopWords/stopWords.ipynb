```python
import spacy
```

```python
from spacy.lang.en.stop_words import STOP_WORDS
print(STOP_WORDS)
len(STOP_WORDS)

nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp("We just opened our wings, the flying part will be coming soon!")

for token in doc:
    if token.is_stop:
        print(token)
```

```python
def preprocess(text):
    doc = nlp(text)

    no_stop_words = [token.text for token in doc if not token.is_stop and not token.is_punct]

    # for token in doc:
    #     if token.is_stop:
    #         no_stop_words.append(token.text)
    
    return (" ").join(no_stop_words)
```

```python
preprocess("We have just opened our wings, the flying part is about to happen!")
```

```python
preprocess("The other is not another, but your divine brother")
```

```python
df = pd.read_json("doj_press.json", lines=True)
df.head()
```

```python
df["contents_new"] = df["contents"].apply(preprocess)
df.head()
```

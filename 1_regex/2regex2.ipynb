```python
text='''
Born	Elon Reeve Musk
June 28, 1971 (age 50)
Pretoria, Transvaal, South Africa
Citizenship	
South Africa (1971â€“present)
Canada (1971â€“present)
United States (2002â€“present)
Education	University of Pennsylvania (BS, BA)
Title	
Founder, CEO and Chief Engineer of SpaceX
CEO and product architect of Tesla, Inc.
Founder of The Boring Company and X.com (now part of PayPal)
Co-founder of Neuralink, OpenAI, and Zip2
Spouse(s)	
Justine Wilson
â€‹
â€‹(m. 2000; div. 2008)â€‹
Talulah Riley
â€‹
â€‹(m. 2010; div. 2012)â€‹
â€‹
â€‹(m. 2013; div. 2016)
'''

pattern4 = r'age (\d+)'
age = re.findall(pattern4, text)
age

pattern5 = r'Born(.*)\n'
name = re.findall(pattern5, text)
name

pattern6 = r'Born.*\n(.*)\(age'
dob = re.findall(pattern6, text)
dob

pattern7 = r'age.*\n(.*)'
pob = re.findall(pattern7, text)
pob
```

```python
def get_pattern_match(pattern, text):
    matches = re.findall(pattern, text)
    if matches:
        return matches[0]
```

```python
get_pattern_match(pattern7, text)
```

```python
def get_personal_information(text):
    name = get_pattern_match(pattern5, text)
    age = get_pattern_match(pattern4, text)
    dob = get_pattern_match(pattern6, text)
    pob = get_pattern_match(pattern7, text)

    return {
        'name': name.strip(),
        'age': int(age),
        'dob':dob.strip(),
        'pob':pob.strip()
    }

get_personal_information(text)
```

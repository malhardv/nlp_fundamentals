```python
import spacy
```

```python
nlp = spacy.blank("en")
```

```python
text = '''
Look for data to help you address the question. Governments are good sources because data from public research is often freely available. Good
places to start include http://www.data.gov/, and http://www.science.gov/, and in the United Kingdom, http://data.gov.uk/.
'''

doc = nlp(text)
urls = []

for token in doc:
    if(token.like_url):
        urls.append(token.text)

urls
```

```python
transactions = "Tony gave 2 $ to Peter, Bruce gave 500 € to Bruce."
doc = nlp(transactions)
money = []

for token in doc:
    if(token.is_currency):
        value = doc[token.i - 1]
        money.append((value.text, token.text))

money
```

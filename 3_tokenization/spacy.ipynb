```python
nlp = spacy.blank("hi")   # hi --> hindi

doc = nlp("मेरे पास 10000 ₹ है, और मैंने कल 5000 ₹  जीते !")
for token in doc:
    print(token, token.is_currency, token.like_num)
```

```python
doc = nlp("gimme double cheese extra large healthy pizza")

tokens = [token.text for token in doc]
tokens
```

```python
from spacy.symbols import ORTH

nlp.tokenizer.add_special_case("gimme", [
    # {ORTH: "give"},   this will give error because we cannot modify the string, so split it into gim and me
    # {ORTH: "me"}
    {ORTH : "gim"},
    {ORTH : "me"}
])

doc = nlp("gimme double cheese extra large healthy pizza")

tokens = [token.text for token in doc]
tokens
```

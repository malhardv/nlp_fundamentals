`Stemming` - Using certain rules to reduce the word, for eg. removing -ing, -able, etc, not necessarily gives a meaningful word in the end.
running - runn
talking - talk
ability - abil
remarkable - remark

`Lemmatization` - Using linguistic knowledge to root down the word to its base word.
running - run
ate - eat
ability - ability

```python
import nltk
```

```python
from nltk.stem import PorterStemmer, SnowballStemmer

stemmer = PorterStemmer()
```

```python
words = ["eating", "eats", "eat", "ate", "adjustable", "rafting", "ability", "meeting"]

for word in words:
    print(word, " | ", stemmer.stem(word))
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
doc = nlp("Mando talked for 3 hours although talking isn't really his thing.")

for token in doc:
    print(token, " | ", token.lemma_, " | ", token.lemma)   # lemma_ returns the root word, lemma returns the hashed word
```

```python
nlp.pipe_names
```

```python
doc = nlp("Bro, you wanna go? Bruh, don't say no! I am exhausted")

for token in doc:
    print(token, " | ", token.lemma_, " | ")

# bro and bruh return to bro and bruh, not rooting to brother
```

```python
# we use attribute ruler to change some rules
ar = nlp.get_pipe('attribute_ruler')

ar.add([[{"TEXT": "Bro"}], [{"TEXT": "Bruh"}]], {"LEMMA": "Brother"}) # is case sensitive
doc = nlp("Bro, you wanna go? Bruh, don't say no! I am exhausted")

for token in doc:
    print(token, " | ", token.lemma_, " | ")
```
